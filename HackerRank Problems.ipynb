```python
import textwrap

def wrap(string, max_width):
    # Use textwrap.wrap() with the provided max_width
    wrapped_lines = textwrap.wrap(string, max_width)
    
    # Join the wrapped lines into a single string using '\n' as the separator
    wrapped_string = '\n'.join(wrapped_lines)
    
    # Return the wrapped string
    return wrapped_string

# Example usage:
wrapped_text = wrap("This is a long string that we want to wrap within a certain width to make it easier to read.", 20)
print(wrapped_text)
```

input:

# any()

This expression returns True if any element of the iterable is true.
If the iterable is empty, it will return False.

```python
# any()

oh = any([1>0,1==0,1<0])

acha = any([1<0,2<1,3<2])

print(oh)
print(acha)
```

# all()

This expression returns True if all of the elements of the iterable are true. If the iterable is empty, it will return True.


```python

ha = all(['a'<'b','b'<'c'])
print(ha)
thike = all(['a'<'b','c'<'b'])
print(thike)
```

<!-- # remove x: Removes element x from the set. If x is not present in the set, it raises a KeyError.

# discard x: Removes element x from the set if it is present. If x is not present, it does nothing.

# pop: Removes and returns an arbitrary element from the set. If the set is empty, it raises a KeyError. -->

```python
#You have a non-empty set , and you have to execute  commands given in  lines.The commands will be pop, remove and discard.

n = int(input())

s = set(map(int,input().split()))

# Number of commands

num_commands = int(input())

for i in range(num_commands):
    command = input().split()
    operation = command[0]
    if operation == "remove":
        s.remove(int(command[1]))
    elif operation == "discard":
        s.discard(int(command[1]))
    elif operation == "pop":
        s.pop()
        # Removes an arbitrary element from the set.Since sets are unordered, pop removes the first element when converted to a list (typically the smallest element).In this case, pop() will remove 1.

print(sum(s))
print(s)
```

# Mini-Max Sum

```python
def minMaxSum(arr):
    arr = sorted(arr)
    min_sum = sum(arr[:len(arr)-1])
    max_sum = sum(arr[1:])
    return min_sum, max_sum
arr=[1, 3, 5, 7, 9]
minMaxSum(arr)
```
