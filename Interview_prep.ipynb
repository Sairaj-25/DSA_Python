# operator overloading

Answer: Operator overloading allows you to define custom behaviors for operators in user-defined classes. You can implement this by defining magic methods like add, sub, eq.

How Operator Overloading Works:

Python provides special methods (also called magic methods or dunder methods, because they start and end with double underscores) that you can override in your class to define the behavior of operators.

For example:

1. The + operator uses the __add__() method.
2. The - operator uses the __sub__() method.
3. The * operator uses the __mul__() method.
4. The == operator uses the __eq__() method.


Explanation:


__init__ Method:

This is the constructor method that initializes the x and y coordinates of the Point object.


__add__ Method:

This is the magic method that overloads the + operator.
When p1 + p2 is executed, the __add__ method is called with p1 as self and p2 as other.
Inside __add__, we check if other is an instance of the Point class. If it is, we return a new Point object with the sum of the corresponding coordinates.
If other is not a Point, we return NotImplemented, which allows Python to try the reverse operation (other + self) or raise a TypeError.
 

__repr__ Method:

This magic method defines how the Point object is represented as a string. It’s used when you print the object or inspect it in an interactive session.
In this case, Point(6, 8) is the output when print(p3) is called.


How It Works:
When you add p1 and p2 using p1 + p2, Python internally calls p1.__add__(p2).
The __add__ method creates and returns a new Point object with the sum of p1's and p2's coordinates.
The result, p3, is a Point object with x = 6 and y = 8.

# GIL (Global Interpreter Lock)? 

### Q. What is GIL (Global Interpreter Lock)?

The Global Interpreter Lock (GIL) is a mutex(or lock) (mutual exclusion) that protects access to Python objects and ensures that only one thread executes Python bytecode at a time, even in a multi-threaded program. 

This is necessary because Python’s memory management is not thread-safe.

In simple terms, the GIL limits Python's ability to use multiple threads for parallel execution of Python code. While threads can be created, only one thread can execute Python code at a time, which can lead to performance bottlenecks in CPU-bound applications (those that require heavy computation).

However, the GIL does not affect I/O-bound tasks (e.g., file reading, writing, or network communication), where threads are often waiting for resources rather than doing computations.


### Q  What is Multi-threading?

Multi-threading is a technique that allows multiple threads (smaller units of a process) to run concurrently within a single process. Threads can share memory and resources, which makes inter-thread communication easy, but it also increases the complexity of managing shared resources.

In Python, multi-threading is often used for I/O-bound tasks rather than CPU-bound tasks due to the GIL. For CPU-bound tasks, the multiprocessing module is often preferred to achieve parallelism.


### Q How do you achieve parallelism in Python, given the GIL?

Answer: For true parallelism in CPU-bound tasks, the multiprocessing module is used, which creates separate processes with their own memory space, bypassing the GIL.


* What is the difference between multi-threading and multi-processing in Python?

Answer: Multi-threading runs multiple threads in the same process, which share memory and resources but are limited by the GIL. Multi-processing, on the other hand, runs multiple processes, each with its own memory space, allowing true parallelism.


### Q  What is Multiprocessing in Python?

Multiprocessing in Python allows the concurrent execution of multiple processes, enabling parallelism and better utilization of CPU cores. Unlike multi-threading, where the Global Interpreter Lock (GIL) restricts the execution of Python threads, multiprocessing bypasses the GIL by using separate processes, each with its own memory space and interpreter instance. This allows true parallelism, especially for CPU-bound tasks (tasks that require significant CPU time).

# Why use Multiprocessing?

Overcome the GIL: Multiprocessing creates separate memory spaces for each process, allowing true parallelism on multi-core processors.

CPU-bound tasks: For tasks that require heavy computations (like numerical simulations, video processing, etc.), multiprocessing can significantly speed up execution.

Scaling to multiple cores: Multiprocessing lets you take full advantage of multiple cores in your CPU by running independent processes.

# How to achieve multiprocessing in Python?
Python provides the multiprocessing module to create and manage multiple processes. Each process runs independently and has its own memory space.

```python
class Point:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __add__(self, other):
        if isinstance(other, Point):
            print(f"Point({self.x} + {other.x}, {self.y} + {other.y})")
            return Point(self.x + other.x, self.y + other.y)     
        return NotImplemented

    def __repr__(self):
        return f"Point({self.x}, {self.y})"

# Creating two Point objects
p1 = Point(2, 3)
p2 = Point(4, 5)

# Adding two Point objects using the overloaded + operator
p3 = p1 + p2

print(p3)  # Output: Point(6, 8)
```

```python
from collections import Counter
def canFormPalindrome(S):
    # Count the frequency of each character
    freq = Counter(S)
    
    # Initialize odd frequency counter
    odd_count = 0
    
    # Count odd frequencies using a for loop
    for count in freq.values():
        if count % 2 != 0:
            odd_count += 1
    
    # A string can form a palindrome if at most one character has an odd frequency
     
    if odd_count <= 1:
        return 1
    else:
        return 0

# Example Usage
S = "geeksogeeks"
canFormPalindrome(S)
# print("Yes" if result == 1 else "No")  # Output: Yes
```

```python
# FInd first non repetating Character

def first_non_repeating_character(s):
    count = {}
    for char in s:
        count[char] = count.get(char, 0) + 1
    for char in s:
        if count[char] == 1:
            return char
    return None

print(first_non_repeating_character("swiss"))
```

```python
def missingNumber(arr):
    arr = sorted(arr)  # Sort the array
    mn, mx = min(arr), max(arr)  # Minimum and maximum values
    expected_sum = (mx * (mx + 1)) // 2 - ((mn - 1) * mn) // 2  # Sum of range [mn, mx+1]
    actual_sum = sum(arr)  # Sum of array elements
    missing = expected_sum - actual_sum  # Missing number
    
    # Handle edge cases where the missing number is outside the range
    if missing == 0:
        if arr[0] == 1:
            return mx + 1
        else:
            return mn - 1
    return missing

# Example Usage
arr = [11, 12, 14, 15]
print(missingNumber(arr))  # Output: 13
```

```python
# Longest common prefix

def long_prefix(strs):
    if not strs:
        return ""
    prefix = strs[0]
    for s in strs[1:]:
        while s[:len(prefix)] != prefix:
            prefix = prefix[:-1]
            if not prefix:
                return None
    return prefix
 
print(long_prefix(["flower", "flow", "float"]))
```

```python
# generate panthesis

def generate_parentheses(n):
    result = []
    def backtrack(s='', left=0, right=0):
        if len(s) == 2 * n:
            result.append(s)
            return
        if left < n:
            backtrack(s + '(', left+1, right)
        if right < left:
            backtrack(s + ')', left, right+1)
    backtrack()
    return result

print(generate_parentheses(3))
```

```python
# Multiprocessing with CPU-bound tasks

import multiprocessing
import time

def square_numbers(numbers):
    for n in numbers:
        time.sleep(0.2)  # Simulating a time-consuming task
        print(f'Square: {n * n}')

if __name__ == '__main__':
    numbers = [1, 2, 3, 4, 5]

    # Creating multiple processes to perform the same task on different data
    process1 = multiprocessing.Process(target=square_numbers, args=(numbers,))
    process2 = multiprocessing.Process(target=square_numbers, args=(numbers,))

    # Start the processes
    process1.start()
    process2.start()

    # Wait for both processes to finish
    process1.join()
    process2.join()

    print("Completed multiprocessing.")

square_numbers(numbers)
```
